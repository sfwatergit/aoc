```python
# Part 1
from functools import cache
from collections import deque
from math import prod
import math
import re

@cache
def get_op(op,num,old):
    if num == "old":
        num = old
    return op([num,old])
    


class Monkey:
    def __init__(self, starting_items, operation, test):
        self.items = deque(starting_items)
        self.operation = operation
        self.test = test
        self.count=0
        
    def inspect_next_item(self):
        self.count+=1
        return self.items.popleft()
    @cache
    def perform_test(self, item):
        if item%self.test[0]==0:
            return self.test[1]
        else:
            return self.test[2]
    
test_path = "test.txt"
input_path = "../input/11.txt"
with open(input_path) as f:
    monkeys = []
    test = []
    for line in f.readlines():
        if line.startswith("  Starting"):
            _,items=line.strip("\n").split(":")
            starting_items = items.split(", ")
        elif line.startswith("  Operation"):
            operation = line.strip("\n")
            _,s = operation.split(": ")
            _,s = s.split("= old ")
            op,num = s.split(" ")
            if num == "old":
                num = "old"
            else:
                num = int(num)
            if op == "*":
                op = prod
            else:
                op = sum
            operation = [op, num]
        elif line.startswith("  Test") or line.startswith("    If true") or line.startswith("    If false"):
            x=list(re.finditer(r"[\d]+",line,re.MULTILINE))[0].group()
            test.append(int(x))
        elif line == "\n":
            m = Monkey(starting_items,operation,test)
            monkeys.append(m)
            test = []


for i in range(20):
    for idx,monkey in enumerate(monkeys):
        while len(monkey.items)!=0:
            cur_worry = int(monkey.inspect_next_item())
            cur_worry=get_op(monkey.operation[0],monkey.operation[1], cur_worry)
            cur_worry=int(cur_worry/3)
            nm=monkey.perform_test(cur_worry)
            monkeys[nm].items.append(cur_worry)


print("Part 1:", prod(sorted([monkey.count for monkey in monkeys])[-2::]))

```

```python
# Part 1

@cache
def get_op(op,num,old):
    if num == "old":
        num = old
    return op([num,old])
    


class Monkey:
    def __init__(self, starting_items, operation, test):
        self.items = deque(starting_items)
        self.operation = operation
        self.test = test
        self.count=0
        
    def inspect_next_item(self):
        self.count+=1
        return self.items.popleft()
    @cache
    def perform_test(self, item):
        if item%self.test[0]==0:
            return self.test[1]
        else:
            return self.test[2]
    
test_path = "test.txt"
input_path = "../input/11.txt"
with open(input_path) as f:
    monkeys = []
    test = []
    for line in f.readlines():
        if line.startswith("  Starting"):
            _,items=line.strip("\n").split(":")
            starting_items = items.split(", ")
        elif line.startswith("  Operation"):
            operation = line.strip("\n")
            _,s = operation.split(": ")
            _,s = s.split("= old ")
            op,num = s.split(" ")
            if num == "old":
                num = "old"
            else:
                num = int(num)
            if op == "*":
                op = prod
            else:
                op = sum
            operation = [op, num]
        elif line.startswith("  Test") or line.startswith("    If true") or line.startswith("    If false"):
            x=list(re.finditer(r"[\d]+",line,re.MULTILINE))[0].group()
            test.append(int(x))
        elif line == "\n":
            m = Monkey(starting_items,operation,test)
            monkeys.append(m)
            test = []

mod = math.prod(monkey.test[0] for monkey in monkeys)

for i in range(10000):
    for idx,monkey in enumerate(monkeys):
        while len(monkey.items)!=0:
            cur_worry = int(monkey.inspect_next_item())
            cur_worry=get_op(monkey.operation[0],monkey.operation[1], cur_worry)
            cur_worry = cur_worry%mod
            nm=monkey.perform_test(cur_worry)
            monkeys[nm].items.append(cur_worry)
print("Part 2:", prod(sorted([monkey.count for monkey in monkeys])[-2::]))
```
