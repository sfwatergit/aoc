```python
from typing import T
import heapq

class PriorityQueue:
    def __init__(self):
        self.elements: list[tuple[float, T]] = []
    
    def empty(self) -> bool:
        return not self.elements
    
    def put(self, item: T, priority: float):
        heapq.heappush(self.elements, (priority, item))
    
    def get(self) -> T:
        return heapq.heappop(self.elements)[1]
```

```python
from collections import deque

test_path = "test.txt"
input_path = "../input/12.txt"

grid = []

def conv(l):
    if l.isupper():
        return l
    else:
        return ord(l)-97



with open(input_path, 'r') as f:
    for r,line in enumerate(f.readlines()):
        row = []
        for c,l in enumerate(line.strip("\n")):
            if l == "E":
                E = (r,c)
                l = 26
            elif l == "S":
                S = (r,c)
                l = 0
            else:
                l = ord(l)-97
            row.append(l)
        grid.append(row)


dirs = {'^':(1,0),"v":(-1,0),"<":(0,-1),">":(0,1)}

frontier = PriorityQueue()
frontier.put(S,0)
distance = {}
distance[S] = 0
came_from = {}
came_from[S]=None

while not frontier.empty():
    curr = frontier.get()
    if curr == E:
        break
    for dy,dx in dirs.values():
        ny,nx = curr[0]+dy,curr[1]+dx
        if nx<0 or ny<0 or nx>=len(grid[0]) or ny>=len(grid):
            continue
        else:
#             print(frontier.elements)
            np = (ny,nx)
            if (grid[ny][nx]-grid[curr[0]][curr[1]])>1:
                continue
            new_cost = distance[curr]+1
            if np not in distance or new_cost<distance[np]:
                frontier.put(np,new_cost)
                distance[np]=new_cost
                came_from[np]=curr
distance
```

```python
import collections
import string
 
inputFile = open(input_path, 'r')
 
rows = [list(row) for row in inputFile.read().split("\n")]
 
elevation = {v: i for i, v in enumerate(string.ascii_lowercase)}
 
def distance(start):
    visited = set([start])
    q = collections.deque([(start[0], start[1], 0)])
 
    while q:
        i, j, d = q.popleft()
        
        if rows[i][j] == "E":
            return d 
 
        for ni, nj in (i + 1, j), (i - 1, j), (i, j - 1), (i, j + 1):
            if 0 <= ni < len(rows) and 0 <= nj < len(rows[0]): 
                if rows[ni][nj] == "E" and elevation["z"] - elevation[rows[i][j]] <= 1:
                    q.append((ni, nj, d + 1))
                elif rows[ni][nj] != "E" and (ni, nj) not in visited and elevation[rows[ni][nj]] - elevation[rows[i][j]] <= 1:
                    visited.add((ni, nj))
                    q.append((ni, nj, d + 1))
 
 
# part 1:
for i in range(len(rows)):
    for j in range(len(rows[i])):
        if rows[i][j] == "S":
            rows[i][j] = "a"
            print(distance((i, j)))
 
# part 2
res = float("inf")
for i in range(len(rows)):
    for j in range(len(rows[i])):
        if rows[i][j] == "S" or rows[i][j] == "a":
            rows[i][j] = "a"
            d = distance((i, j))
            if d:
                res = min(res, d)            
print(res)
 
```

```python
from typing import *

Location = TypeVar('Location')

def reconstruct_path(came_from: dict[Location, Location],
                     start: Location, goal: Location) -> list[Location]:

    current: Location = goal
    path: list[Location] = []
    if goal not in came_from: # no path was found
        return []
    while current != start:
        path.append(current)
        current = came_from[current]
    path.append(start) # optional
    path.reverse() # optional
    return path
```
