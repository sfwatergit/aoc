```python
dirs_to_sum = []
dirs_to_min = []
fs = {"TOTAL":0}
# with open("test.txt",'r') as f:
with open("../input/7.txt",'r') as f:
    f.readline()
    line = f.readline().strip("\n")
    cur_dir = fs
    dirs = []
    while line != "":
        
        if line.startswith("$"):
            _, command = line.strip("\n").split(" ",1)
            if command.startswith("cd"):
                _ , target = command.strip("\n").split(" ")
                if target == "..":
                    cur_total = cur_dir["TOTAL"]
                    if cur_total<100000:
                        dirs_to_sum.append(cur_dir["TOTAL"])
                    cur_dir = dirs.pop()
                    cur_dir["TOTAL"]+=cur_total
                else:
                    dirs.append(cur_dir)
                    cur_dir = cur_dir[target]
            elif command.startswith("ls"):
                pass
        elif line.startswith("dir"):
            _, dn = line.strip("\n").split(" ")
            cur_dir[dn] = {"TOTAL": 0}
        else: # line starts with number, so file
            size, _ = line.strip("\n").split(" ")
            cur_dir["TOTAL"] += int(size)
        line = f.readline()
print("Part 1:", sum(dirs_to_sum))
dirs_to_min
```

```python
def list_sizes(fs,sizes):
    for k,v in fs.items():
        
        print(k,v)
        if isinstance(v,int):
            sizes.append(v)
        elif isinstance(v,dict):
            list_sizes(v,sizes)
    return sizes
```

```python
cur_dir = []
dirs = {} # [fullname]->[sz,parent_dir]
for l in open("../input/7.txt","rt"):
  e = l.strip().split()
  if l.startswith("$ cd"):
    if e[2] == "..":
      cur_dir.pop()
    else:
      cur_dir.append(e[2])
    p = '/'.join(cur_dir[:-1])
    d = '/'.join(cur_dir)
    if d not in dirs:
      dirs[d] = [0,p]
  elif e[0] not in ("$","dir"):
    s = int(e[0])
    d = '/'.join(cur_dir)
    while d:
      dirs[d][0] += s
      d = dirs[d][1]
t = 0
d = []
for s,_ in dirs.values():
  d.append(s)
  if s <= 100_000:
    t += s
```
