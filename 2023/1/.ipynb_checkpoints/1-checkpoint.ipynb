```python
# Part 1
import re


cur_tot = 0
with open("../input/1.txt", 'r') as f:
    for line in f.readlines():
        line = line.strip("\n")
        numbers =  [int(s) for s in line if s.isdigit()]
        s1 = str(numbers[0])
        s2 = str(numbers[-1])
        dig=s1+s2
        cur_tot += int(dig)
        
cur_tot
```

```python
# Part 2

cur_tot = 0

digits = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine", "ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen", "sixteen"]

with open("../input/1.txt", 'r') as f:
    for line in f.readlines():
        line = line.strip("\n")
        # get first digit
        lefts = ""
        for l in line:
            if l.isdigit():
                first = l
                break
            else:
                lefts+=l
            
            for digit in digits:
                if digit in lefts:

                    first = str(digits.index(digit)+1)
                    if len(first)>1:
                        first = first[-1]
                    break
            if first is not None:
                break
        # get last digit
        rights = ""
        for l in reversed(line):
            if l.isdigit():
                last = l
                break
            else:
                rights=l+rights
            for digit in digits:
                if digit in rights:

                    last = str(digits.index(digit)+1)
                    if len(last)>1:
                        last = last[-1]
                    break
            if last is not None:
                break

        dig = first+last
        first = None
        last = None

        cur_tot+=int(dig)
                
print(cur_tot)
```
