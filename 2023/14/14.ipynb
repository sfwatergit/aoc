```python
# Part 2
from functools import cache

# inp = "test.txt"
inp = "../input/14.txt"


with open(inp) as f:
    grid = [l.strip("\n") for l in f.readlines()]

grid = np.array([[el for el in l] for l in grid])

n_rows = len(grid)
n_cols = len(grid[0])

R = "O"
RR = "#"
AIR = "."
total = 0


def mr_north(grid):
    grid = grid.copy()
    for r in range(n_rows-1,0,-1):
        for c in range(n_cols):
            el = grid[r][c]
            if el == R: #
                up_el = grid[r-1][c]
                if up_el == AIR:
                    grid[r-1][c] = R
                    grid[r][c] = AIR
    return grid

def mr_west(grid):
    grid = grid.copy()
    for r in range(n_rows):
        for c in range(n_cols-1,0,-1):
            el = grid[r][c]
            if el == R: #
                up_el = grid[r][c-1]
                if up_el == AIR:
                    grid[r][c-1] = R
                    grid[r][c] = AIR
    return grid

def mr_south(grid):
    grid = grid.copy()
    for r in range(0,n_rows-1):
        for c in range(n_cols):
            el = grid[r][c]
            if el == R: #
                up_el = grid[r+1][c]
                if up_el == AIR:
                    grid[r+1][c] = R
                    grid[r][c] = AIR
    return grid


def mr_east(grid):
    grid = grid.copy()
    for r in range(n_rows):
        for c in range(0,n_cols-1):
            el = grid[r][c]
            if el == R: #
                up_el = grid[r][c+1]
                if up_el == AIR:
                    grid[r][c+1] = R
                    grid[r][c] = AIR
    return grid
    

def move_rocks(grid,fn):
    prev_grid=grid
    while True:
        grid = fn(prev_grid)
        if np.array_equal(grid,prev_grid):
            break
        prev_grid = grid
    return grid

totals = []
states = {}
cycles = 1_000_000_000
for c in range(200):
    total= 0
    grid=move_rocks(grid, mr_north)    
    grid=move_rocks(grid, mr_west)
    grid=move_rocks(grid, mr_south)
    grid=move_rocks(grid, mr_east)
    cur_state = "".join(''.join(x) for x in grid)
    if cur_state in states:
        loop_size = c-states[cur_state]
        loop_start = states[cur_state]
    else:
        states[cur_state]=c
    
step = (cycles - (loop_start+1)) % loop_size

for c in range(step):
    total= 0
    grid=move_rocks(grid, mr_north)    
    grid=move_rocks(grid, mr_west)
    grid=move_rocks(grid, mr_south)
    grid=move_rocks(grid, mr_east)
        
for r in range(n_rows):
    for c in range(n_cols):
        if grid[r][c] == R:
            total+=(n_rows-r)
total
```
