```python
# Part 1

# inp = "test.txt"
inp = "../input/15.txt"

with open(inp) as f:
    steps = f.read().split(",")

    
def hash_algo(step):
    cur_value = 0
    for char in step:
        cur_value+=ord(char)
        cur_value*=17
        cur_value%=256
    return cur_value

sum([hash_algo(step) for step in steps])
```

```python
# Part 2
import re
from collections import OrderedDict
from math import prod

inp = "test.txt"
inp = "../input/15.txt"

pattern = re.compile("([a-zA-Z]+)([=|-])(\d?)")
with open(inp) as f:
    steps = f.read().split(",")
    
    
class LensBox(OrderedDict):
    'Store items in the order the keys were last added'
    
    def __init__(self,box_no):
        super().__init__()
        self.box_no=box_no

    def __setitem__(self, key, value):
        super().__setitem__(key, value)

    
    def __repr__(self):
        if len(self.keys())>0:
            s = str(list(self.values()))
            return  f"Box {self.box_no}: {s}"
        else:
            return ""
    
    def focusing_power(self):
        if len(self.values())>0:
            return sum([(self.box_no+1)*(1+slot)*int(k.split(" ")[1]) for slot,k in enumerate(self.values())])
        else:
            return 0
    
boxes = [LensBox(idx) for idx in range(256)]

for step in steps:
#     print(step)
    label,op,f = pattern.match(step).groups()
    box_no = hash_algo(label)
    lens = f"{label} {f}"
    if op == '=':
        if label in boxes[box_no].keys():
            boxes[box_no][label]=lens
        else:
            boxes[box_no][label]=lens
            boxes[box_no].move_to_end(label)
    if op == '-':
        box = boxes[box_no]
        if label in box.keys():
            del box[label]
#     print(boxes)
#     print("\n")

sum([box.focusing_power() for box in boxes])
```
