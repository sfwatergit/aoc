```python
import heapq
import itertools
import enum
import collections





inp = "../input/7.txt"
# inp = "test.txt"

with open(inp) as f:
    lines = [line.strip("\n").split(" ") for line in f.readlines()]

HandType = enum.Enum('HandType',['FiveOaK','FourOaK','FullHouse','ThreeOaK',"TwoPair","OnePair","HighCard"])
    
```

```python
from functools import cmp_to_key

def determine_hand_type(hand):
    c = collections.Counter(hand).most_common()
    if c[0][1]==5:
        return HandType.FiveOaK
    elif c[0][1]==4:
        return HandType.FourOaK
    elif c[0][1]==3:
        if c[1][1]==2:
            return HandType.FullHouse
        else: 
            return HandType.ThreeOaK
    elif c[0][1]==2:
        if c[1][1]==2:
            return HandType.TwoPair
        else: 
            return HandType.OnePair
    else: 
        return HandType.HighCard

CARD_TO_NUM_MAP = {'T':10,'J':11,'Q':12,'K':13,'A':14}

def card_to_num(card): 
    if card.isnumeric():
        return int(card)
    else:
        return CARD_TO_NUM_MAP[card]


def compare_in_group(self, other):
    self,other = self[1][0],other[1][0]
    for i in range(5):
        if self[i] != other[i]:
            return -1 if card_to_num(self[i]) < card_to_num(other[i]) else 1
    return 0
    

# Part 1:

els = []
idx = itertools.count(1)
for _,h_list in sorted([(i,sorted(k,key=cmp_to_key(compare_in_group))) for i,k in itertools.groupby(sorted([(determine_hand_type(hand),(hand,bid)) for hand,bid in lines], key=lambda k: k[0].value),key=lambda x:x[0].value)],reverse=True):
    els.append([int(h[1][1])*next(idx) for h in h_list])
sum(list(itertools.chain(*els)))
```

```python
# Part 2:

def determine_hand_type(hand):
    c = collections.Counter(hand).most_common()
    num_j = [x for x in c if x[0]=='J']
    
    if len(num_j)>0:
        c.remove(num_j[0])
        if len(c)==0:
            return HandType.FiveOaK
        most=c[0]
        c.remove(most)
        most = [(most[0],most[1]+num_j[0][1])]
        c = most+c
    if c[0][1]==5:
        return HandType.FiveOaK
    elif c[0][1]==4:
        return HandType.FourOaK
    elif c[0][1]==3:
        if c[1][1]==2:
            return HandType.FullHouse
        else: 
            return HandType.ThreeOaK
    elif c[0][1]==2:
        if c[1][1]==2:
            return HandType.TwoPair
        else: 
            return HandType.OnePair
    else: 
        return HandType.HighCard
    
els = []
idx = itertools.count(1)
for _,h_list in sorted([(i,sorted(k,key=cmp_to_key(compare_in_group))) for i,k in itertools.groupby(sorted([(determine_hand_type(hand),(hand,bid)) for hand,bid in lines], key=lambda k: k[0].value),key=lambda x:x[0].value)],reverse=True):
    els.append([int(h[1][1])*next(idx) for h in h_list])
sum(list(itertools.chain(*els)))
```
